# SWARMPLOT

<b>So it is our second categorical scatterplot- SWARMPLOT. Now in stripplot points are overlapping. In swarmplot points are non-overlapping. But for large number of datatset it does not work perfectly always.<br>Now to make swarmplot the command is sns.swarmplot().</b>

```python
import matplotlib.pyplot as plt         #importing packages
import seaborn as sns
import pandas as pd
```

```python
mpg=sns.load_dataset("tips")            #loading tips dataset
```

```python
mpg.head(6)                             #printing first 6 rows
```

```python
sns.swarmplot(data=mpg.head(100),x='size',y='tip')          #making swarmplot
plt.show()
```

<b>We can pass hue,hue_order,palette,color all parameters. We will use some of these to explore differrent parameters.</b>

```python
color=sns.husl_palette(2)                                                           #setting color palette(see more in seaborn api)
sns.swarmplot(data=mpg.head(100),x='size',y='tip',hue='sex',palette=color,dodge=True)          #making swarmplot using hue,palette,dodge
plt.show()
```

<b>Now we will use parameter - size,marker,alpha.</b>

```python
color=sns.husl_palette(2)                                                           #setting color palette(see more in seaborn api)
sns.swarmplot(data=mpg.head(100),x='size',y='tip',hue='sex',palette=color,dodge=True,marker="D",alpha=0.6,size=4)       #making swarmplot using size,marker,alpha
plt.show()
```
