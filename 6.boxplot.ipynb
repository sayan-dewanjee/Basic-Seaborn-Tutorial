# BOXPLOT

<b>So now we will make Boxplot. Here the command is sns.boxplot() . Now we already know about most of the parameters -hue,hue_order,palette,orient etc. We will discuss more about the new parameters.</b>

```python
import matplotlib.pyplot as plt               #importing packages
import seaborn as sns
import pandas as pd
```

```python
mpg=sns.load_dataset("tips")                  #loading dataset
```

```python
mpg.head(5)                                   #printing first 5 rows
```

```python
sns.boxplot(data=mpg,x='day',y='total_bill',hue="sex",palette=sns.husl_palette(n_colors=2,s=5),hue_order=['Female','Male'])
plt.show()
```

<b>Now here we will understand orient parameter.Notice the following 2 plots : </b>

```python
sns.boxplot(data=mpg,y='size',x='total_bill',hue="sex")
plt.show()
```

<b>As here both axis takes numerical value here boxplot is made through total_bill as default. But here categorical variable is size. So we want to make boxplot about that. So we will now change orientation from vertical to horizontal</b>

```python
sns.boxplot(data=mpg,y='size',x='total_bill',hue="sex",orient="h")   #changing the orientation
plt.show()
```

```python
sns.boxplot(data=mpg,x='size',y='total_bill',hue="sex")   #both these plots are same.Here we interchange x and y. basically seaborn take x axis as
#default to make boxplot.Changing orientation chnage boxplots vertically or horizontally.
plt.show()
```

<b>Now we will explore saturation and fill parameter. Saturation parameter determines the saturation of filled color of boxes. fill parameter takes boolean value and determines whether box will be filled or not.</b>

```python
sns.boxplot(data=mpg,y='size',x='total_bill',hue="sex",orient="y",saturation=0.1)   #using saturation parameter
plt.show()
```

```python
sns.boxplot(data=mpg,y='size',x='total_bill',hue="sex",orient="h",fill=False)   #using fill parameter
plt.show()
```

<b>Now we will explore width,gap,linewidth and whis parameter. Width determines width of boxes,gap determines gaps between boxes,whis determines length of whiskers,linewidth determines of width of border line.</b>

```python
sns.boxplot(data=mpg,x='size',y='total_bill',hue="sex",width=0.3,gap=0.2,linewidth=00.8)       #using width and gap parameter
plt.show()
```

<b>Another parameter is fliersize.Basically fliersize determines the size of outliers. And wemcan change fliers too. but for this we have to use function of matplotlib. flierprops={"marker":"D"} or flierprops={"marker":"x"} are commands to do this. Inplace of D and x we can use any vali marker style to customize it.</b>

```python
sns.boxplot(data=mpg,x='size',y='total_bill',hue="sex",width=0.3,gap=0.2,fliersize=8,flierprops={"marker":"D"})       #using fliersize parameter
plt.show()
```

```python
sns.boxplot(data=mpg,x='size',y='total_bill',hue="sex",width=0.3,gap=0.2,flierprops={"marker":">"})       #using fliersize parameter
plt.show()
```
