# Scatterplot

<b>At first we will import packages and load our crab dataset.Then we will draw scatterplot</b>

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
mpg=pd.read_csv("C:\\Users\\SAYAN DEWANJEE\\Documents\\python\\crab.csv") #importing csv file
```

```python
mpg #printing dataframe
```

<b>Now we want to make a scatterplot between Length and Height. We will use sns.scatterplot command</b>

```python
sns.scatterplot(data=mpg,x="Length",y="Height")
plt.show()                                      #Making scatterplot
```

<b>Now we can similarly set hue,style and size parameter.</b>

```python
sns.scatterplot(data=mpg.head(100),x="Length",y="Height",hue="Sex",style="Sex")  #setting sex as hue and style parameter
plt.show()              
```

```python
sns.scatterplot(data=mpg.head(100),x="Length",y="Height",hue="Sex",size="Sex")  #setting sex as hue and size parameter
plt.show()  
```

```python
sns.scatterplot(data=mpg.head(100),x="Length",y="Height",hue="Sex",size="Sex",style='Sex',sizes=[40,42,48])  #setting sex as hue,style and size parameter
plt.show()   #Note we are correcting sizes too here.otherwise green dots are too small
```
