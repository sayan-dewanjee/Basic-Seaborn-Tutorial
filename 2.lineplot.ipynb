# LinePlot

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
mpg=pd.read_csv("crab.csv") #Loading dataset
```

```python
mpg #showing dataset
```

```python
mp=mpg.head(10) #taking only first 10 values for lineplot
```

```python
mp #showing the reduced dataset
```

<b>So at first we import our required packages. Then we select crab.csv dataset for plotting.Now we reduced this large dataset by considering first 20 values.Now we will proceed with lineplot</b>

```python
sns.lineplot(data=mp,x="Shucked_Weight",y="Weight") #to plot simple line plot
```

<b>Note here an additional comment has appeared. To remove this we use show() function</b>

```python
sns.lineplot(data=mp,x="Shucked_Weight",y="Weight")
plt.show() #Removing additional comments
```

<b>Now we will understand hue,size and style parameter by applying.</b>

```python
sns.lineplot(data=mp,x="Shucked_Weight",y="Weight",hue="Sex")
plt.show() #Using hue parameter
```

```python
sns.lineplot(data=mp,x="Shucked_Weight",y="Weight",size="Sex")
plt.show() #Using size parameter
```

```python
sns.lineplot(data=mp,x="Shucked_Weight",y="Weight",style="Sex")
plt.show() #Using style parameter
```

<b>So basically hue,style and size works similarly. But hue classifies categories with differrent colour,size classifies categories with differrent line width,style classifies categories with differrent line shape.We can add hue,shape and style in single command.</b>

```python
sns.lineplot(data=mp,x="Shucked_Weight",y="Weight",style="Sex",hue="Sex")
plt.show() #Using hue and style parameter
```

<b>So even we can use hue and style parameter in same statement. Hue and style parameter need not be same. So using hue and style parameter we can classify our visualizations in 2 more categories</b>

<b>Now we want to explore hue parameter more. We will use palette command to change colour of hue categories.</b>

```python
sns.lineplot(data=mp,x="Shucked_Weight",y="Weight",hue="Sex",palette="magma")
plt.show() #Using palette parameter. Here palette name is magma
```

```python
sns.lineplot(data=mp,x="Shucked_Weight",y="Weight",hue="Sex",palette="tab10")
plt.show() #It is another palette named tab10
```

<b>Now we want to explore hue_order parameter. So this parameter sets in which order hue will work. Note here categories are F,M,I repectively. Now we will change it to M,F,I</b>

```python
sns.lineplot(data=mp,x="Shucked_Weight",y="Weight",hue="Sex",hue_order=['M','F','I'])
plt.show() #Using hue_order parameter
```

<b>Now we want to explore size parameter. We have 2 more parameters - sizes and size_order which works with size parameter</b>

```python
sns.lineplot(data=mp,x="Shucked_Weight",y="Weight",size="Sex",sizes=(1,3))
plt.show() #Using sizes parameter
```

<b>So we can see sizes parameter basically changes the size of the linewidth of our size category. We can pass a list to set differrent width for differrent categories,</b>

```python
sns.lineplot(data=mp,x="Shucked_Weight",y="Weight",size="Sex",sizes=[5,1,3])
plt.show() #Passing list into sizes parameter
```

<b>size_order is a same parameter like hue_order. It passes in which order size_order will make categories.</b>

<b>Now we will explore style parameter.We have 3 differrent parameters to use with style parameter. They are dashes,markers and style_order. Now style-order is same  as hue_order and size_order. Now we will explore dashes parameter.</b>

<b>dashes parameter takes boolean values or list values. If we put dashes=True it will use default dashes or we can pass a list.</b>

```python
sns.lineplot(data=mp,x="Shucked_Weight",y="Weight",style="Sex",dashes=[(2,1),(3,1),(4,1)])
plt.show() #Using dashes parameter
```

<b>So we can passes list to dashes parameter. Basically (2,1) means we will get a line of 2 unit and then a gap of 1 unit. Similarly (3,1) means we will get a line of 3 units and then gap of 1 units.</b>

<b>Now we will learn about markers. Markers basically draw the corner points with special charachter.Markers take boolean or list value like dashes.Setting it True will result in default markers. Otherwise we can pass a list</b>

```python
sns.lineplot(data=mp,x="Shucked_Weight",y="Weight",style="Sex",markers=True)
plt.show() #Using style parameter
```

```python
sns.lineplot(data=mp,x="Shucked_Weight",y="Weight",style="Sex",markers=[8,9,10])
plt.show() #Using style parameter
```

<b>So we passes a list of possible markers and corner points are marked with those markers. Now the list of markers can be accessed from here : https://matplotlib.org/stable/api/markers_api.html </b>

<b>Now we almost completed hue,size and style parameter. We will use in almost all plots in seaborn to classify categories.Now we will see some general parameter of line plot</b>

```python
sns.lineplot(data=mp,x="Shucked_Weight",y="Weight",sort=False)
plt.show() #Using sort parameter
```

<b>So here we use sort parameter. It takes boolean values - True and False. Setting it true data will be sorted by the x and y variables. Setting it false will connect points as they appear in dataset</b>

<b>Another parameter is legend. It takes 4 keywords-â€œautoâ€, â€œbriefâ€, â€œfullâ€, or False. False will vanish legends and auto will choose whether brief or full legend will be used. Full legend implies that all legends will be shown and brief legend implies that a summary will be shown.Here we have very few category hence here brief and full will make same output.</b>
