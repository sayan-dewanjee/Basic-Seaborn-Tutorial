<b>Now we will move to categorical plot. The general format of this type plot is catplot. Under catplot we have 8 differrent type plots. Those are:-      
- Categorical scatter plot - stripplot,swarmplot
- Categorical distribution plot - boxplot,violinplot,boxenplot
- Categorical estimate plot - pointplot,barplot,countplot

All these plots can be accessed by sns.catplot and kind parameter{Previously we see relplot and by changing kind parameter we were getting 2 differrent plots}. Here we will not discuss boxenplot and pointplot.</b>

<b>Now we will see stripplot by using command sns.stripplot(). Before that we will load tips dataset here again.</b>

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd        #importing packages
```

```python
mpg=sns.load_dataset('tips')  #importing dataset
```

```python
mpg.head(5)   #showing first 5 rows
```

```python
sns.stripplot(data=mpg,x="day")  #here we can pass x and y individually 
plt.show()
```

```python
sns.stripplot(data=mpg,y="tip",x='day')  #here we can pass x and y both
plt.show()
```

<b>So we pass one categorical variable and another numerical variable. Stripplot is best for making such plots. Though we can pass both numerical data or both categorical data.</b>

<b>We can pass hue and hue_order parameter here similarly.And we can pass order parameter to specify order of categories of categorical variable.</b>

```python
sns.stripplot(data=mpg,y="tip",x='day',hue='smoker',order=['Sun','Sat','Thur','Fri'])  # using hue and order parameter
plt.show()
```

<b>Now we can see here points are overlapping and it makes our plot complicated. To remove this we use jitter parameter. We can pass True to use default value. Otherwise we can pass value to make plot clear to our extent</b>

```python
sns.stripplot(data=mpg,y="tip",x='day',hue='smoker',jitter=0.3)  # using jitter parameter
plt.show()
```

<b>If we do not use jitter parameter it keeps our points randomly. And when we pass a value it will maintain that value. Now let's see what happens when we pass jitter=False</b>

```python
sns.stripplot(data=mpg,y="tip",x='day',hue='smoker',jitter=False)  # making jitter parameter false
plt.show()
```

<b>Now if we want to see smoker categories clearly,we have to use dodge parameter.</b>

```python
sns.stripplot(data=mpg,y="tip",x='day',hue='smoker',jitter=0.3,dodge=True)  # using dodge parameter
plt.show()
```

<b>We can pass size,edgecolor and linewidth parameter. Size determines size of points. edgecolor determines edgecolor of points and linewidth determines width of lines that frame plot elements.</b>

```python
sns.stripplot(data=mpg,y="tip",x='day',hue='smoker',jitter=0.3,dodge=True,size=3)  # using size  parameter
plt.show()
```

```python
sns.stripplot(data=mpg,y="tip",x='day',hue='smoker',jitter=0.3,dodge=True,linewidth=2)  # using linewidth  parameter
plt.show()
```

<b>Now we will discuss about 2 differrent scales log_scale and native_scale.Basically log_scale=True changes our scale of numerical axis into logarithmic scale.We will show use of native_scale in following example.  </b>

```python
sns.stripplot(data=mpg,x="tip",y='total_bill') #when we are not using native_scale
plt.show()
```

<b>Here x axis is not working properly as tip is numerical value. Now we will pass native_scale to clear it.</b>

```python
sns.stripplot(data=mpg,x="tip",y='total_bill',native_scale=True)  #using native_scale
plt.show()
```

<b>Another important parameter is orient. It takes 2 values - 'v','h' . 'v' implies vertical,'h' implies horizontal. It is useful when both axis takes numerical variable. It passes in which variable we will make grouping. We will use it in future in barplot and many other plots.</b>

<b>Even we can pass marker parameter here also to make our plot more beautiful. another parameter is alpha which controls the opacity of our points.</b>

```python
sns.stripplot(data=mpg,y="tip",x='day',hue='smoker',jitter=False,dodge=True,marker="D",alpha=0.5)  # using marker parameter
plt.show()
```

```python
sns.stripplot(data=mpg,y="tip",x='day',hue='smoker',jitter=False,marker="D",alpha=0.3,dodge=True)  # decreasing alpha parameter
plt.show()
```
