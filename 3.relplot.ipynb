# RelPlot

<b>Seaborn Relplot is a special type of relational plot between two variable x and y. Before this, we have seen 2 differrent relational plot-scatter plot and line plot. RelPlot provides these 2 plots with more additional feature.
To see differrent aspects of RelPlot we will load tips dataset which is already loaded into seaborn.</b>

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
tips=sns.load_dataset("tips") #loading tips dataset
```

```python
tips.head(10) #loading first 10 rows
```

<b>At first we use the basic sns.relplot command to plot the relation</b>

```python
sns.relplot(data=tips,x='total_bill',y='tip') #using relplot command
plt.show()
```

```python
sns.relplot(data=tips,x='total_bill',y='tip',kind='line') #using kind parameter
plt.show()
```

<b>So if we pass basic relplot command it makes scatterplot. And in second plot we use kind parameter. kind parameter takes two values - line or scatter. So from relplot command we can make scatter or line plot. and like previous we can pass hue,style and size parameter.</b>

```python
sns.relplot(data=tips,x='total_bill',y='tip',kind='scatter',hue='sex',style='smoker',palette='magma',markers=[8,'x'])
plt.show()                     #using hue and style parameter
```

<b>But relplot has more additional features.We will now explore them. Here additonal parameters are row,row_order,col,col_order.
If we use hue,style,size parameter it will make our plot very complicated.Now row parameter makes differrent plots for a differrent levels of a variable.</b>

```python
sns.relplot(data=tips,x='total_bill',y='tip',row="sex") #using row command
plt.show()
```

<b>So here we get 2 differrent plot for male and female and rowwise.Now we will provide row_order parameter.</b>

```python
sns.relplot(data=tips,x='total_bill',y='tip',row="sex",row_order=['Female','Male']) #using row_order command
plt.show()
```

<b>Now we will provide col parameter.</b>

```python
sns.relplot(data=tips,x='total_bill',y='tip',col="sex") #using col command
plt.show()
```

<b>Now we will use row and col both paramter to create 4 plots. With this we will use hue parameter to classify categories.</b>

```python
sns.relplot(data=tips,x='total_bill',y='tip',col="sex",row='smoker',hue='time') #using col and row command
plt.show()
```

<b>Now we will use another parameter col_wrap. This parameter takes an integer value at which it wraps column width.</b>

```python
sns.relplot(data=tips,x='total_bill',y='tip',col="size",hue='time') #using col command
plt.show()
```

<b>We are not being able to see plots clearly. We will wrap column to 3 plots.</b>

```python
sns.relplot(data=tips,x='total_bill',y='tip',col="size",hue='time',col_wrap=3) #using col_wrap command
plt.show()
```

<b>N.B. We will not be able to row with col_wrap parameter</b>

<b>There are 2 final parameters height and aspect. Now we will explore these two parameters.Note here we are not being able to see all the plots and plots are large. So now we will decrease height and aspect of every plot.</b>

```python
sns.relplot(data=tips,x='total_bill',y='tip',col="sex",row='smoker',hue='time',kind='line',height=3,aspect=1) #using height and aspect command
plt.show()
```
